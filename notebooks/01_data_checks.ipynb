# Data Checks

Load processed NPZ files and visualize key statistics.

```python
import numpy as np, pathlib as P
root = P.Path('data/h36m/annotations')
files = list(root.glob('*.npz'))[:3]
files
```

```python
d = np.load(files[0]); print(d.keys()); [a.shape for a in (d['k2d'], d.get('k3d'), d.get('conf'))]
```

```python
import matplotlib.pyplot as plt
k2d = d['k2d']
plt.figure(); plt.hist(k2d.reshape(-1,2)[:,0], bins=50); plt.title('k2d x distribution'); plt.show()
```
